#### 创建一个DataFrame

```python
import pandas as pd
```

```python
df = pd.DataFrame({"a":[1,2,3],"b":[4,5,6],"c":[7,8,9]})
```

```python
df
```

```python
df["b"].at[1]  # df["b"] 是一个Series，此时at函数只需取行名即可
```

```python
df.at[1,"b"]  # at函数通过行名和列名来取值
```

```python
df.iat[1,1]  # iat函数通过行号和列号来取值，行号和列号从0开始。
```

```python
f = pd.DataFrame({"name":["aoe","bob","cool"],"value":[3,4,5],"nature":["a","b","c"]})
```

```python
f.set_index("name",inplace = True,drop = False)
```

```python
f
```

```python
f.at["bob","nature"]
```

```python
f["value"].at["cool"]
```

```python
f["test"]=None
```

```python
for i in f.index:
    f["test"].at[i] = f["value"].at[i]+1
```

```python
f
```

```python
f["test1"]=f["value"]+1
```

```python
def getsum(x):
    return x["a"]+x["b"]+x["c"]
```

```python
df["d"]=df.apply(getsum,axis=1)
```

```python
df
```

```python
df["e"]=df.apply(lambda x:x["a"]+x["b"],axis=1) # 相当于df["e"]=df["a"]+df["b"],但apply能实现更复杂的计算
```
